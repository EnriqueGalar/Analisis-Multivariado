#Ejercicio 4

```python
#Reescalamos
bank_tilde=bank.copy()
bank_tilde[0]=bank_tilde[0]/10
bank_tilde[1]=bank_tilde[1]/10
bank_tilde[2]=bank_tilde[2]/10
bank_tilde[5]=bank_tilde[5]/10
```

```python
#Matriz de unos
ones=np.ones(200)
#medias=(1/n)X.T (unos)
vec_medias=(1/200)*(bank_tilde.values.T @ ones)
#Sigma=(1/n)X.T(X)-medias por medias trans. Mejor usamos el método de Numpy
Sigma=np.cov(bank_tilde.T.values)
#Calcular eigenvalores y eigenvectores
evalue, evect = np.linalg.eig(Sigma)
```

```python
vec_medias
```

```python
evalue
```

```python
evect
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
principalComponents = pca.fit_transform(bank_tilde.values)
principalDf = pd.DataFrame(data = principalComponents, columns = ['PC1', 'PC2', 'PC3'])
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Componente Principal 1', fontsize = 15)
ax.set_ylabel('Componente Principal 2', fontsize = 15)
ax.set_title('Proyección de 2 Componentes Principales', fontsize = 20)
targets = ['Verdaderos', 'Falsos']
ax.scatter(principalDf['PC1'][:100].values
               , principalDf['PC2'][:100].values
               , c = '#0CECDD'
               , s = 50)
ax.scatter(principalDf['PC1'][100:].values
               , principalDf['PC2'][100:].values
               , c = '#C400FF'
               , s = 50)
ax.legend(targets)
ax.grid()
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Componente Principal 2', fontsize = 15)
ax.set_ylabel('Componente Principal 3', fontsize = 15)
ax.set_title('Proyección de 2 Componentes Principales', fontsize = 20)
targets = ['Verdaderos', 'Falsos']
ax.scatter(principalDf['PC2'][:100].values
               , principalDf['PC3'][:100].values
               , c = '#0CECDD'
               , s = 50)
ax.scatter(principalDf['PC2'][100:].values
               , principalDf['PC3'][100:].values
               , c = '#C400FF'
               , s = 50)
ax.legend(targets)
ax.grid()
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Componente Principal 1', fontsize = 15)
ax.set_ylabel('Componente Principal 3', fontsize = 15)
ax.set_title('Proyección de 2 Componentes Principales', fontsize = 20)
targets = ['Verdaderos', 'Falsos']
ax.scatter(principalDf['PC1'][:100].values
               , principalDf['PC3'][:100].values
               , c = '#0CECDD'
               , s = 50)
ax.scatter(principalDf['PC1'][100:].values
               , principalDf['PC3'][100:].values
               , c = '#C400FF'
               , s = 50)
ax.legend(targets)
ax.grid()
```

```python
data=df.drop([0],axis=1)
```

```python
#Matriz de unos
ones_=np.ones(23)
#medias=(1/n)X.T (unos)
medias_=(1/23)*(data.values.T @ ones_)
#Sigma=(1/n)X.T(X)-medias por medias trans. Mejor usamos el método de Numpy
Sigma_=np.cov(data.T.values)
#Calcular eigenvalores y eigenvectores
evalue_, evect_ = np.linalg.eig(Sigma_)
```
